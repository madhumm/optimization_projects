```python
import math 
import numpy as np
```

```python
def f(x):
    return (math.exp(x) - x)
```

```python
def f_prime(x):
    return math.exp(x) - 1
```

```python
def f_double_prime(x):
    return math.exp(x)
```

```python
print(f(-1))
print(f_prime(-1))
print(f_double_prime(-1))

print(10**-5)
```

```python
x_0 = -1

x_array = [None] * 100
x_array[0] = -1

f_x_k = f(-1)

print(f_x_k)

f_prime_x_k = f_prime(-1)

print(f_prime_x_k)

index = 1

while abs(f_prime_x_k) > 10**-5:
    
    cur_x_val = x_array[index-1]
    
    print('curr x_k value : {0}'.format(cur_x_val))
    
    derivative_factor = (f_prime(cur_x_val)/f_double_prime(cur_x_val))
    
    x_k = x_array[index-1] - derivative_factor
    
    print('new x_k value : {0}'.format(x_k))
    
    x_array[index] = x_k
    
    print('f(x) value : {0}'.format(f(x_k)))
    
    f_prime_x_k = f_prime(x_k)
    
    print('f_prime(x) value : {0}'.format(f_prime_x_k))    
    
    index +=1
    
    print('Index value : {0}'.format(index))
    
    
    
```

```python
import matplotlib.pyplot as plt
import numpy as np
  
feature_x = np.linspace(-10.0, 12.0, 2)
feature_y = np.linspace(-10.0, 12.0, 2)
  
# Creating 2-D grid of features
[x, y] = np.meshgrid(feature_x, feature_y)
  
fig, ax = plt.subplots(1, 1)
  
g_partial_1 = np.sin(x)*np.sin(y)
    
g_partial_2 = np.exp(abs(100 - (np.sqrt(x**2 + y**2)/np.pi)))

Z = -0.0001 * (abs(g_partial_1 * g_partial_2)+ 1)**0.1
  
# plots filled contour plot
ax.contourf(x, y, Z)
  
ax.set_title('Filled Contour Plot')
ax.set_xlabel('feature_x')
ax.set_ylabel('feature_y')
  
plt.show()
```

```python
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import random

def fun(x, y):
    return x + y

def f(x, y):
    return (1- x + x*y)**2 + (2 - x + x**2*y)**2 + (3 - x + x**3 *y )**2


fig = plt.figure(figsize=(8,5))
ax = fig.add_subplot(111, projection='3d')
n = 10
xs = [i for i in range(-5, 5, 1) for _ in range(-5, 6, 1)]
ys = list(range(-5, 6, 1)) * n
zs = [f(x, y) for x,y in zip(xs,ys)]

ax.scatter(xs, ys, zs)

plot = ax.plot_surface(X=xs, Y=ys, Z=zs, cmap='YlGnBu_r', vmin=0, vmax=200)

ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')

plt.show()
```

```python
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import random

def fun(x, y):
    return x**2 + y

def f(x, y):
    return (1- x + x*y)**2 + (2 - x + x**2*y)**2 + (3 - x + x**3 *y )**2

fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111, projection='3d')
x = y = np.arange(-6.0, 6.0, 0.05)
X, Y = np.meshgrid(x, y)
zs = np.array([f(x,y) for x,y in zip(np.ravel(X), np.ravel(Y))])
Z = zs.reshape(X.shape)

ax.plot_surface(X, Y, Z)

ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')

plt.show()
```
